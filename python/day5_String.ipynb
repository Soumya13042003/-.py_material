## String

String are used to store text data in python. Any value written inside single qoutes or double qoutes is treated as string.

* Creating String : Strings are can be created using single qoutes or double qoutes.
                a = 'abc'
                b = "def"

* Multiline String: Created using triple qoutes.
                a = ''' this is 
                        a multiple
                        strings '''
