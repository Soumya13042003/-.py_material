## Tuple

Tuple are ordered collections of items that are immutable. They are similar to list, but their immutability makes them different.

Key: 
* Ordered : Items have a defined, unchangable order.
* Immutable : Once a tuple is created, you cannot change add or remove item.
* Duplicates : Because tuples are indexed, they can contain multiple items with the same value
* Heterogeneous : tuples can store elements of different data types with in structure.
     
Example: tuple = ('a', "b", "c")
         print(tuple)

```python
## creating a tuple
a = (1,2,3,'helloworld')
print(a)
print(type(a))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
tpl = tuple([1, 2, 3, 4, 5])
print(tpl, type(tpl))
```

```python
## Immutable Nature Of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned.

lst=(1,2,3,4,5)
print(lst)
```

```python
lst[1]="hello world"
lst
```

```python
## packing and unpacking tuple
## Packing means storing multiple values into a single tuple.

a = 1, 2, 3.14, 'hello world'
print(a)
```

```python
## Unpacking means extracting tuple values into variables.
b, c, d, e = a
print(b)
print(c)
print(d)
print(e)
```

```python
## Nested Tuple
## Nested List

lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
print(lst[0][0:3])
print(lst[2][0:3])
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
