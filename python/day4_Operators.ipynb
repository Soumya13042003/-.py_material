## Operators
Operators in general are used to perform operations are values and variable.

There are five types of operator:

- Airthmetic Operators- +, -, *, /, %, //, **
- Relational Operators- <, =< , > , =>, == , =!
- Logical Operators- AND, OR, NOT
- Bitwaise Operators- &, ^, ~, <<, >>
- Assignment Operators- =, +=, -=, *=, %=

* #### Airthmetic Operators
Python airthmetic operators are used to perform basic mathematic operations like algorithm like addition, substraction, multiplication and division.   

```python
a = 5
b = 10

addition = a + b #simple addition


print(addition)

print(type(addition))
```

```python
a = 5.5
b = 10

addition = a + b #simple addition


print(addition)

print(type(addition))
```

```python
a = 5
b = 10

substraction = a - b #simple subtraction


print(substraction)

print(type(substraction))
```

```python
#a basic calculator program

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b) # Returns only the left-over remainder
print("Exponentiation:", a ** b) ## means a raised to the power of b
print("Floor Division:", a // b) ## Division rounded down to integer
```

## Logical Operator

Logical Operators performs logical AND, OR, NOT operations. It is used to combine conditions statement.

```python
## And ,Not,OR
X=True
Y=True

result = X and Y
print(result)
```

```python
X=False
Y=True

result =X and Y
print(result)
```

```python
## OR
X=False
Y=False

result =X or Y
print(result)
```

```python
# Not operator
X=False
not X
```
