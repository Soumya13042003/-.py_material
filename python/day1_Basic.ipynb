Key Characters:

1. Readability: Means Python code is written in a way that is easy to understand and locks similar to english language.

2. Interpreated language: Python is an Interpreated language, which means code is executed line by without compilation. Errors are shown immediately when the interpreter reaches that line.

3. Dynamically Typed: Python you don't need to declare the data type of a variable. The type is decided automatically at runtime.

4. Indentation: Indentation in python refers to the spaces at the beginning of a line used to define the structre and blocks of code.

Key: 1. Python uses Indentation instead of {}
     2. Usually 4 spaces or single tab are used.
     3. All statements in a blocks must have the same Indentation.

```python
## Indentation
##Python uses indentation to define blocks of code. Consistent use of spaces (commonly 4) or a tab is required.

age=32
if age>30:
    print(age)
print(age)
```

```python
### Simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
